```python
# Data Visualization

This notebook demonstrates essential data visualization techniques used to understand data before applying machine learning models.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
## Sample Dataset

We use a small sample dataset to demonstrate different visualization techniques.
```

```python
data = {
    "Age": [22, 25, 30, 35, 40, 28, 32],
    "Salary": [30000, 40000, 50000, 60000, 70000, 45000, 52000],
    "Department": ["AI", "ML", "Data", "AI", "Data", "ML", "AI"]
}

df = pd.DataFrame(data)
df
```

```python
## Line Plot

Line plots are useful for visualizing trends over a continuous variable.
```

```python
plt.plot(df["Age"], df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary Trend")
plt.show()
```

```python
## Bar Plot

Bar plots compare values across different categories.
```

```python
sns.barplot(x="Department", y="Salary", data=df)
plt.title("Average Salary by Department")
plt.show()
```

```python
## Histogram

Histograms show the distribution of a single variable.
```

```python
plt.hist(df["Age"], bins=5)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
## Scatter Plot

Scatter plots visualize the relationship between two numerical variables.
```

```python
plt.scatter(df["Age"], df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.show()
```

```python
## Box Plot

Box plots help identify outliers and data spread.
```

```python
sns.boxplot(x="Department", y="Salary", data=df)
plt.title("Salary Distribution by Department")
plt.show()
```

```python
## Heatmap

Heatmaps visualize correlations between numerical features.
```

```python
corr = df[["Age", "Salary"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
## Summary

This notebook demonstrated common visualization techniques including line plots, bar charts, histograms, scatter plots, box plots, and heatmaps for data analysis.
```
